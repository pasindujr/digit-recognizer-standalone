## Scaling data

```python
# The MNIST dataset has images with pixel values in the range [0, 255]
X = X / 255
X_test = X_test / 255
```

```python
X
```

### Split data for test and training data

```python
from sklearn.model_selection import train_test_split

X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.1, random_state=1)
X_train.shape, X_val.shape, y_train.shape, y_val.shape, X_test.shape
```

### Defining the model and check against training data

```python
from sklearn.neural_network import MLPClassifier

model = MLPClassifier(
    hidden_layer_sizes=(100, 100),  # Two hidden layers with 100 neurons each
    activation='relu',             # Rectified Linear Unit activation function
    solver='adam',                 # Optimization algorithm
    alpha=0.0001,                  # L2 penalty (regularization) term
    batch_size='auto',             # Size of minibatches for optimization
    learning_rate='constant',      # Learning rate schedule
    learning_rate_init=0.001,      # Initial learning rate
    max_iter=200,                  # Maximum number of iterations
    random_state=1                 # Random seed for reproducibility
)



# Fit the model to the training data
model.fit(X_train, y_train)

# Now you can use the score method to evaluate the model's performance
print(model.score(X_train, y_train))
```

### Get predictions against validation data and print the accuracy

```python
y_pred = model.predict(X_val)
print(model.score(X_val, y_val))
```

```python
df_val = pd.DataFrame({'actual_value':y_val, 'predicted_value':y_pred})
df_val.reset_index(drop=True, inplace=True)
df_val
```

## Creating the pickle file

```python
from PIL import Image
import requests
import io
import joblib

# Load and preprocess a single image
def preprocess_single_image(image_url):
    response = requests.get(image_url)
    image_data = response.content
    threshold = 128

    # Create a PIL Image object from the downloaded image data
    img = Image.open(io.BytesIO(image_data))

    # Resize the image to 28x28 pixels
    img = img.resize((28, 28))

    # Convert the image to grayscale
    img = img.convert('L')

    # Apply thresholding to binarize the image
    img = img.point(lambda p: p < threshold and 255)

    # Flatten the binarized image into a 1D array of 784 binary pixel values
    img_array = np.array(img).flatten() / 255.0  # Normalize to [0, 1]

    return img_array

# Replace 'your_single_image.jpg' with the path to your single image
image_path = 'https://i.ibb.co/sHCvBVD/2023-09-02-0hl-Kleki.png'
single_image = preprocess_single_image(image_path)

# Make a prediction on the single image
prediction = model.predict([single_image])


# Print the predicted label
print(f'Predicted label: {prediction[0]}')
joblib.dump(model, 'model_filename.pkl')
```
